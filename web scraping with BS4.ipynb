```python
#importing libraries
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
# creating list for url from page 1 to 300
url_list = []
page_count = list(range(1, 301))
for i in page_count:
    url= "https://www.jagran.com/latest-news-page"+str(i)+".html"
    url_list.append(url)
```

```python
print(url_list)
```

```python
# getting data from website and storing it in list
title_list = []
content_list = []
for url in url_list:
    r = requests.get(url)
    soup = bs(r.content)
    data_con = soup.find_all("div", class_="protxt fr")
    for i in data_con:
        title = i.find("div", class_="h3")
        content = i.find("p")
        title_list.append(title.text)
        content_list.append(content.text)
```

```python
print(content_list)
```

```python
#saving data using pandas
dictionary = {"Title":title_list, "Content":content_list}
content_table = pd.DataFrame(dictionary)
content_table.to_csv('title_content_table.csv')
```
